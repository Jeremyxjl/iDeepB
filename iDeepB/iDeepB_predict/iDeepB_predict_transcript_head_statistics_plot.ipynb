```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_table("iDeepB_train_val_metric.txt", header=None, sep="\t")
data.columns = ["protein", "cell", "loss_fun", "log", "loss", "pearson", "spearman", "auc", "prauc", "mse"]
```

```python
data.head(1)
```

```python
data['loss'] = data['loss'].str.replace('loss=', '')
data['pearson'] = data['pearson'].str.replace('pearson=', '')
data['spearman'] = data['spearman'].str.replace('spearman=', '')
data['auc'] = data['auc'].str.replace('auc=', '')
data['prauc'] = data['prauc'].str.replace('prauc=', '')
data.replace('nan', np.nan, inplace=True)
```
